## LZ77

```python
def LZ77_encode(mensaje,S=12,L=6):
    output = []
    search_buffer = ''
    dic = ''
    i = 0
    while i < len(mensaje):
        if i == 0:
            output.append((1,0,mensaje[0]))
            search_buffer = mensaje[1:1+L]
            dic = mensaje[0]
            i += 1
            continue
        posicion = 0
        longitud_maxima_cadena = 0
        simbolo = search_buffer[0]
        search = dic + search_buffer
        j = len(dic) - 1
        while j >= 0:
            if dic[j] == search_buffer[0]:
                l = 1
                while len(dic)+l < len(search):
                    if search[j+l] == search[len(dic)+l]:
                        l +=1
                    else:
                        break
                if l > longitud_maxima_cadena:
                    posicion = len(dic) - j
                    longitud_maxima_cadena = l
                    if len(dic)+l < len(search):
                        simbolo = search[len(dic)+l]
                    elif i+longitud_maxima_cadena < len(search):
                        simbolo = search[i+longitud_maxima_cadena]
                    else:
                        simbolo = search[len(dic)+l-1]
                        longitud_maxima_cadena -= 1
            j -= 1
        i += longitud_maxima_cadena+1
        dic = mensaje[i-S if i-S >= 0 else 0:i]
        search_buffer = mensaje[i:i+L+1]
        if longitud_maxima_cadena == 0:
            posicion = 1
        output.append((posicion, longitud_maxima_cadena, simbolo))
    return output
```

```python
def LZ77_decode(tok):
    txt = []
    for (pos, length, a) in tok:
        if (length == 0):
            txt.append(a)
        else:
            aux = len(txt)-pos
            for i in range(0,length):
                txt.append(txt[len(txt)-pos])
            txt.append(a)
        
    return ''.join(txt)
```

```python
def LZSS_encode(txt,mm,s,t):
    tok = []
    i = 0
    n = len(txt)
    while i < n:
        j = max(i - s, 0)
        max_match_len = 0
        max_match_pos = 0
        while(j < i):
            match_len = 0
            while i + match_len < n and txt[j + match_len] == txt[i + match_len] and match_len < t:
                match_len += 1
            if match_len >= max_match_len and match_len >= mm:
                max_match_len = match_len
                max_match_pos = i - j
            j += 1 
        if max_match_len == 0:
            tok.append(txt[i])
            i += 1
        else: 
            tok.append((max_match_pos, max_match_len))
        i += max_match_len
    return tok
```

## LZ78

```python
def LZ78_encode(txt):
    tok = []
    tok.append((0, txt[0]))
    dictionary = ['', txt[0]]
    done = False
    current_pos = 1
    while current_pos < len(txt):
        pointer = ''
        for i,x in enumerate(dictionary):
            if x == txt[current_pos]:
                pointer = i
                break
        if not pointer:
            tok.append((0, txt[current_pos]))
            dictionary.append(txt[current_pos])
            current_pos += 1
        else: 
            pos = current_pos+1
            b = True
            while b and pos < len(txt) - 1:
                b = False
                for i,x in enumerate(dictionary):
                    if x == txt[current_pos:pos+1]:
                        pointer = i
                        b = True
                        break
                pos += 1
            if (pos == len(txt) - 1): pos += 1
            tok.append((pointer, txt[pos-1]))
            dictionary.append(txt[current_pos:pos])
            current_pos = pos 

    return tok
```

```python
def LZ78_decode(encoded):
    dictionary = {0: ''}
    next_index = 1
    output = []
    for index, symbol in encoded:
        if symbol:
            if index in dictionary:
                entry = dictionary[index] + symbol
            else:
                entry = dictionary[index - 1] + symbol
            dictionary[next_index] = entry
            next_index += 1
            output.append(entry)
        else:
            output.append(dictionary[index])
    return ''.join(output)
```

```python
def initialize_dictionary(txt):
    chars = list(set(txt))
    chars.sort()
    dictionary = {}
    for i in range(len(chars)):
        dictionary[chars[i]] = i
    return dictionary


def LZW_encode(txt):
    #dictionary = {chr(i): i for i in range(256)}
    dictionary = initialize_dictionary(txt)
    result = []
    
    current_string = ""
    next_code = len(dictionary)
    
    for char in txt:
        test_string = current_string + char
        if test_string in dictionary:
            current_string = test_string
        else:
            result.append(dictionary[current_string])
            dictionary[test_string] = next_code
            next_code += 1
            current_string = char
    
    if current_string:
        result.append(dictionary[current_string])
    
    return result
```

```python
def LZW_decode(codigos, alfabeto):
    diccionario = {i: c for i, c in enumerate(alfabeto)}
    entrada = alfabeto[codigos[0]]
    texto = [entrada]
    for codigo in codigos[1:]:
        if codigo in diccionario:
            cadena = diccionario[codigo]
        elif codigo == len(diccionario):
            cadena = entrada + entrada[0]
        else:
            raise ValueError("Código inválido: {}".format(codigo))
        texto.append(cadena)
        diccionario[len(diccionario)] = entrada + cadena[0]
        entrada = cadena
    return "".join(texto)
```

## Rearrangement of letters

```python
def permutations(txt):
    ans = [txt]
    i = 1
    while i < len(txt):
        ans += [ans[-1][-1] + ans[-1][0:-1]]
        i += 1
    return ans
```

```python
def encode_burrows_wheeler(txt):
    perms = sorted(permutations(txt))
    last_col = ''.join([x[-1] for x in perms])
    pos = perms.index(txt)
    return last_col, pos
```

```python
def decode_burrows_wheeler(last_col, pos):
    n = len(last_col)
    first_col = sorted([(i, x) for i, x in enumerate(last_col)], key=itemgetter(1))
    T = [None for i in range(n)]
    for i, y in enumerate(first_col):
        j, _ = y
        T[j] = i
    Tx = [pos]
    for i in range(1, n):
        Tx.append(T[Tx[i - 1]])
    S = [last_col[i] for i in Tx]
    S.reverse()
    return ''.join(S)
```

```python
def encode_move_to_front(txt):
    alphabet = list(set(txt))
    alphabet.sort()
    q = len(alphabet)
    sequence = []
    for char in txt:
        index = alphabet.index(char)
        sequence.append(index)
        alphabet.pop(index)
        alphabet.insert(0, char)
    return sequence
```

```python
def decode_move_to_front(cod, alp):
    alphabet = list(alp)
    txt = ""
    for index in cod:
        char = alphabet[index]
        txt += char
        alphabet.pop(index)
        alphabet.insert(0, char)
    return txt
```
